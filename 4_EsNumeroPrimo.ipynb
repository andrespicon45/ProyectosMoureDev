Escribe un programa que se encargue de comprobar si un número es o no primo.
Hecho esto, imprime los números primos entre 1 y 100.

```python

try:
    num = int(input('Agrega un numero:'))

except:
    print('Debe ser entero! Intetalo nuevamente.')

def es_primo(num):

    divisores = []

    for i in range(1,num+1):
        if num%i == 0:
                divisores.append(i)

    if len(divisores) > 2:
        print(f'{num} no es un numero primo.')
    else:
        if divisores[len(divisores)-1] != num:
            print('{num} no un numero primo.')
        else:
            print(f'{num} es un numero primo!')

es_primo(num)

print('Te mostrare los numeros primos entre 1 y 100')
for i in range(1,101):
    es_primo(i)
```
